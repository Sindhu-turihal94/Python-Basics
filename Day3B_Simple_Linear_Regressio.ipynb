**Simple Linear Regression**

**Linear Regression** is a basic machine learning algorithm used to **predict a numeric** **value** based on the relationship between variables.



**ðŸ§  Formula:**

y= mx+c

Where:


*  y: predicted output (e.g., salary)

*   x: input feature (e.g., years of experience)
*  m: slope (how much
 increases per unit increase in
)


*  c: intercept (value of
 when x=0
)






Load and inspect data


```python
import pandas as pd
df = pd.read_csv( "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv"
)
df.head()
```

Check for null values


```python
df.isnull().sum()
```

```python
df.info()
```

One-Hot Encode Categorical Variables


```python
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.head()
```

Feature and Target Separation


```python
x = df_encoded.drop("charges", axis=1)
y = df_encoded["charges"]
```

Feature Scaling using StandardScaler


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

**Train-Test Split**

**Explanation**


*   train_test_split() divides the data:

*   X_train, y_train: used to train the model

*  X_test, y_test: used to test the model
*   Parameters:


*  test_size=0.2: 20% of data goes to test set.


*   random_state=42: ensures repeatable results (random shuffling is fixed).



```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

Train Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

**In a Jupyter environment, please rerun this cell to show the HTML representation or trust the notebook.
On GitHub, the HTML representation is unable to render, please try loading this page with nbviewer.org.**

Make Predictions

```python
y_pred = model.predict(x_test)
```

**Evaluate Model**
**Explanation:**

*  mean_squared_error(y_test, y_pred):

*   Measures average squared error.

*  Closer to 0 is better.
*   r2_score(y_test, y_pred):


*  Tells how much of the variance in target variable is explained by the model.


*   Closer to 1 is better.












```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2_score(y_test, y_pred))
```

Visualize Predictions vs Actual

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred,color='gray')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Charges")
plt.title("Actual vs Predicted Charges")
plt.grid(True)
plt.show()
```
